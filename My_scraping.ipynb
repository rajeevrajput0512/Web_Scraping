```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup as soup
```

```python
url = urlopen("https://www.snapdeal.com/products/mobiles-mobile-phones?utm_term=436977179921_102550322078&_tablets_mobiles_tablets___generic_ftu=&utm_campaign=mobiles_&utm_source=earth_semnonbrand&utm_medium=deals&sort=plrty")
```

```python
content = url.read()
```

```python
souped_url = soup(content,"html.parser")
```

```python
print(souped_url)
```

```python
image = souped_url.find_all("img", class_ = "product-image")
print(image)
```

```python
lister = [i.get('src') or i.get('data-src') for i in image]
```

```python
import requests
```

```python
illa = requests.get(lister[0])
```

```python
with open("snap_scrap.jpg","wb") as file:
    file.write(illa.content)
```
